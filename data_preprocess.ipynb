```python
import matplotlib.pyplot as plt

# Group the data by video title and sum the view and like counts
grouped_data = data.groupby('Video Title').agg({'View Count': 'sum', 'Like Count': 'sum'}).reset_index()

# Sort the data by view count in descending order
sorted_data = grouped_data.sort_values(by='View Count', ascending=False)

# Select the top 20 videos
top_20_data = sorted_data.head(20)

# Create the bar graph
plt.figure(figsize=(12, 6))
plt.bar(top_20_data['Video Title'], top_20_data['View Count'], alpha=0.7, label='View Count', color='yellow')
plt.bar(top_20_data['Video Title'], top_20_data['Like Count'], alpha=0.7, label='Like Count', color='red')
plt.xlabel('Video Title')
plt.ylabel('Counts')
plt.title('Top 20 Videos - View Count and Like Count')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()

# Show the bar graph
plt.show()
```
